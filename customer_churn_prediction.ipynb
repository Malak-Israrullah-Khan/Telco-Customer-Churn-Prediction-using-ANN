```python
X = df_clean.drop('Churn', axis=1)
y = df_clean['Churn']
```

```python
print(X.shape)
print(y.shape)
```

```python
categorical_cols = X.select_dtypes(include=['object']).columns
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns

print("Categorical columns:")
print(categorical_cols)

print("\nNumerical columns:")
print(numerical_cols)
```

```python
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)
```

```python
print(X_encoded.shape)
X_encoded.head()
```

```python
X_encoded = X_encoded.astype(float)
```

```python
X_encoded.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)

print("\nTrain target distribution:")
print(y_train.value_counts(normalize=True) * 100)

print("\nTest target distribution:")
print(y_test.value_counts(normalize=True) * 100)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])
```

```python
import torch

X_train_tensor = torch.tensor(X_train.values, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test.values, dtype=torch.float32)

y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).view(-1, 1)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).view(-1, 1)
```

```python
print(X_train_tensor.shape)
print(y_train_tensor.shape)
print(X_test_tensor.shape)
print(y_test_tensor.shape)
```

```python
from torch.utils.data import TensorDataset, DataLoader

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
input_size = X_train_tensor.shape[1]
print("Input size:", input_size)
```

```python
# ===============================
# 5. Build Deep Learning Model
# ===============================

import torch
import torch.nn as nn

class ChurnMLP(nn.Module):
    def __init__(self, input_size):
        super().__init__()
        
        self.network = nn.Sequential(
            nn.Linear(input_size, 64),
            nn.ReLU(),
            
            nn.Linear(64, 32),
            nn.ReLU(),
            
            nn.Linear(32, 16),
            nn.ReLU(),
            
            nn.Linear(16, 1)
        )
    
    def forward(self, x):
        return self.network(x)


model = ChurnMLP(input_size)

print(model)
```

```python
# ===============================
# 6. Loss Function and Optimizer
# ===============================

import torch.optim as optim

criterion = nn.BCEWithLogitsLoss()

optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# ===============================
# 7. Train the Model
# ===============================

epochs = 50
train_losses = []

for epoch in range(epochs):
    model.train()
    
    epoch_loss = 0
    
    for X_batch, y_batch in train_loader:
        # 1. Clear old gradients
        optimizer.zero_grad()
        
        # 2. Forward pass
        outputs = model(X_batch)
        
        # 3. Calculate loss
        loss = criterion(outputs, y_batch)
        
        # 4. Backpropagation
        loss.backward()
        
        # 5. Update weights
        optimizer.step()
        
        # Add batch loss
        epoch_loss += loss.item()
    
    average_loss = epoch_loss / len(train_loader)
    train_losses.append(average_loss)
    
    if (epoch + 1) % 5 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {average_loss:.4f}")
```

```python
# ===============================
# 8. Plot Training Loss
# ===============================

import matplotlib.pyplot as plt

plt.plot(train_losses)
plt.title("Training Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

```python
# ===============================
# 9. Evaluate the Model
# ===============================

model.eval()

all_preds = []
all_labels = []

with torch.no_grad():
    for X_batch, y_batch in test_loader:
        logits = model(X_batch)
        
        probabilities = torch.sigmoid(logits)
        
        predictions = (probabilities >= 0.5).float()
        
        all_preds.extend(predictions.numpy())
        all_labels.extend(y_batch.numpy())
```

```python
# ===============================
# 10. Metrics
# ===============================

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

accuracy = accuracy_score(all_labels, all_preds)
precision = precision_score(all_labels, all_preds)
recall = recall_score(all_labels, all_preds)
f1 = f1_score(all_labels, all_preds)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

print("\nConfusion Matrix:")
print(confusion_matrix(all_labels, all_preds))

print("\nClassification Report:")
print(classification_report(all_labels, all_preds))
```

```python
# ===============================
# 11. Confusion Matrix Plot
# ===============================

cm = confusion_matrix(all_labels, all_preds)

plt.figure(figsize=(5, 4))
plt.imshow(cm)
plt.title("Confusion Matrix")
plt.colorbar()

plt.xticks([0, 1], ['No Churn', 'Churn'])
plt.yticks([0, 1], ['No Churn', 'Churn'])

plt.xlabel("Predicted")
plt.ylabel("Actual")

for i in range(2):
    for j in range(2):
        plt.text(j, i, cm[i, j], ha='center', va='center')

plt.show()
```

```python
# ===============================
# 12. Improved Deep Learning Model
# ===============================

import torch
import torch.nn as nn
import torch.optim as optim

class ImprovedChurnMLP(nn.Module):
    def __init__(self, input_size):
        super().__init__()
        
        self.network = nn.Sequential(
            nn.Linear(input_size, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Dropout(0.2),
            
            nn.Linear(32, 16),
            nn.ReLU(),
            
            nn.Linear(16, 1)
        )
    
    def forward(self, x):
        return self.network(x)


improved_model = ImprovedChurnMLP(input_size)
print(improved_model)
```

```python
# Count class 0 and class 1 in training data
num_neg = (y_train_tensor == 0).sum()
num_pos = (y_train_tensor == 1).sum()

pos_weight = num_neg / num_pos

print("Number of No Churn samples:", num_neg.item())
print("Number of Churn samples:", num_pos.item())
print("Positive class weight:", pos_weight.item())
```

```python
criterion_improved = nn.BCEWithLogitsLoss(pos_weight=pos_weight)

optimizer_improved = optim.Adam(
    improved_model.parameters(),
    lr=0.001,
    weight_decay=0.0001
)
```

```python
# ===============================
# 13. Train Improved Model
# ===============================

epochs = 50
improved_train_losses = []

for epoch in range(epochs):
    improved_model.train()
    
    epoch_loss = 0
    
    for X_batch, y_batch in train_loader:
        optimizer_improved.zero_grad()
        
        outputs = improved_model(X_batch)
        
        loss = criterion_improved(outputs, y_batch)
        
        loss.backward()
        
        optimizer_improved.step()
        
        epoch_loss += loss.item()
    
    average_loss = epoch_loss / len(train_loader)
    improved_train_losses.append(average_loss)
    
    if (epoch + 1) % 5 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {average_loss:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.plot(improved_train_losses)
plt.title("Improved Model Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

```python
# ===============================
# 14. Evaluate Improved Model
# ===============================

improved_model.eval()

improved_preds = []
improved_labels = []

with torch.no_grad():
    for X_batch, y_batch in test_loader:
        logits = improved_model(X_batch)
        probabilities = torch.sigmoid(logits)
        predictions = (probabilities >= 0.5).float()
        
        improved_preds.extend(predictions.numpy())
        improved_labels.extend(y_batch.numpy())
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

accuracy = accuracy_score(improved_labels, improved_preds)
precision = precision_score(improved_labels, improved_preds)
recall = recall_score(improved_labels, improved_preds)
f1 = f1_score(improved_labels, improved_preds)

print("Improved Model Results")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

print("\nConfusion Matrix:")
cm_improved = confusion_matrix(improved_labels, improved_preds)
print(cm_improved)

print("\nClassification Report:")
print(classification_report(improved_labels, improved_preds))
```

```python
comparison = pd.DataFrame({
    "Model": ["Basic MLP", "Improved MLP"],
    "Accuracy": [0.7839, accuracy],
    "Precision": [0.6122, precision],
    "Recall": [0.5107, recall],
    "F1 Score": [0.5569, f1]
})

comparison
```

```python
# ===============================
# 15. Threshold Tuning
# ===============================

improved_model.eval()

all_probs = []
all_true = []

with torch.no_grad():
    for X_batch, y_batch in test_loader:
        logits = improved_model(X_batch)
        probabilities = torch.sigmoid(logits)
        
        all_probs.extend(probabilities.numpy())
        all_true.extend(y_batch.numpy())
```

```python
import numpy as np

all_probs = np.array(all_probs).flatten()
all_true = np.array(all_true).flatten()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

threshold_results = []

for threshold in [0.3, 0.4, 0.5, 0.6, 0.7]:
    preds = (all_probs >= threshold).astype(int)
    
    acc = accuracy_score(all_true, preds)
    prec = precision_score(all_true, preds)
    rec = recall_score(all_true, preds)
    f1 = f1_score(all_true, preds)
    
    threshold_results.append({
        "Threshold": threshold,
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1 Score": f1
    })

threshold_df = pd.DataFrame(threshold_results)
threshold_df
```

```python
plt.plot(threshold_df['Threshold'], threshold_df['Precision'], marker='o', label='Precision')
plt.plot(threshold_df['Threshold'], threshold_df['Recall'], marker='o', label='Recall')
plt.plot(threshold_df['Threshold'], threshold_df['F1 Score'], marker='o', label='F1 Score')

plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Threshold Tuning')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ===============================
# 16. Final Model Evaluation
# ===============================

final_threshold = 0.6

final_preds = (all_probs >= final_threshold).astype(int)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

final_accuracy = accuracy_score(all_true, final_preds)
final_precision = precision_score(all_true, final_preds)
final_recall = recall_score(all_true, final_preds)
final_f1 = f1_score(all_true, final_preds)
final_cm = confusion_matrix(all_true, final_preds)

print("Final Model Results at Threshold:", final_threshold)
print("Accuracy:", final_accuracy)
print("Precision:", final_precision)
print("Recall:", final_recall)
print("F1 Score:", final_f1)

print("\nConfusion Matrix:")
print(final_cm)

print("\nClassification Report:")
print(classification_report(all_true, final_preds))
```
